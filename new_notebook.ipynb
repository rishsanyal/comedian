```python
import numpy as np
import pandas as pd

joke_info_dataset = pd.read_json('datasets/joke_info.json')

joke_info_dataset.head()
```

```python
def apply_joke_length(row):
    row['joke_len'] = len(row['jokeText'])
    return row
    # return len(row['text'])

joke_info_dataset = joke_info_dataset.apply(apply_joke_length, axis=1)
```
